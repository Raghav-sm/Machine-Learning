## Relational Plot  
* Its for relation between 2 or more variables
* Bivariate Analysis


```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
tips.sample(7)
```

```python
# Scatterplot
sns.scatterplot(data = tips, x = tips['tip'], y = tips['total_bill']);
#Axis level function
```

```python
#figure level function 
sns.relplot(data = tips , x = tips['tip'] , y = tips['total_bill'] , 
            kind = 'scatter' , style = tips['sex'] , hue = tips['time'] ,
            size = tips['size']);
#Similarly in kind we can also use lineplot
```

#### Generally if Time is present in X Axis then Lineplot is prefered
