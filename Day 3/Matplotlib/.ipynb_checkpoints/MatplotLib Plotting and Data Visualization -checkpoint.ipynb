`x = np.linspace(0,10,100) `Initializes variable x and assigns it an array containing 100 equally spaced numbers between 0 and 10, using the NumPy library.

````python
x = np.linspace(0,10,100)
````

````python
fig ,ax = plt.subplots()
ax.plot(x , x**2);# Default is lineplot
````

- `fig , ax = plt.subplots()`: This creates a plot with a figure (`fig`) and an axis (`ax`) for further customization.
- `ax.scatter( x , np.exp(x))`: It plots a scatter graph on the axis (`ax`) using the values from the array `x` and their corresponding exponential values, showing data points on the plot.

````python
#Scatterplot
fig , ax = plt.subplots()
ax.scatter( x , np.exp(x));
````


- 📊 `fig , ax = plt.subplots()`: Prepares a plot with a figure (`fig`) and an axis (`ax`) for customization.

- 📊🏷️ `ax.bar(list(fruits.keys()) , list(fruits.values()))`: Plots a bar chart on the axis (`ax`) using the fruit names as labels on the x-axis, and their quantities on the y-axis, creating a chart for a "Fruit Store."

````python
#using Dictionary to create plots
fruits = { "Apple " : 30,
           "Banana" : 5 ,
           "Orange" : 14 } 
fig , ax = plt.subplots()
ax.bar(list(fruits.keys()) , list(fruits.values())); 
ax.set(title = "Fruit Store" , 
       ylabel = 'Number of fruits',
       xlabel = 'Fruit')
````

- 📊 `fig , ax = plt.subplots()`: Initializes a figure (`fig`) and an axis (`ax`) for a plot.

- 📊📈 `ax.barh(list(fruits.keys()) , list(fruits.values()))`: Creates a horizontal bar chart on the axis (`ax`) using fruit names on the y-axis and their quantities on the x-axis, suitable for visualizing data like "Fruit Store" inventory.

````python
fig , ax = plt.subplots()
ax.barh(list(fruits.keys()) , list(fruits.values()));
````

- 🎲 `x = np.random.randn(1000)`: Generates an array `x` containing 1000 random numbers sampled from a standard normal distribution.

- 📊 `fig , ax = plt.subplots()`: Sets up a plot with a figure (`fig`) and an axis (`ax`) for further customization.

- 📊📊 `ax.hist(x)`: Plots a histogram on the axis (`ax`) to display the distribution of the random numbers generated in `x`.

````python
#histogram
x = np.random.randn(1000)
fig , ax = plt.subplots()
ax.hist( x );
````

- 📊✏️ `fig , ((ax1 ,ax2) , (ax3,ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))`: Sets up a 2x2 grid of subplots within a figure (`fig`) for creating multiple plots.

- 📈📊📊 `ax1.plot(x, x/2)`, `ax2.scatter(...)`, `ax3.bar(...)`, and `ax4.hist(...)`:   Each subplot (`ax1`, `ax2`, `ax3`, `ax4`) is used to plot different types of data, such as a line plot, scatter plot, bar chart, and histogram. This allows you to visualize various data simultaneously in one figure.

````python
# Thr is two options for subplots
#option 1 
fig , ((ax1 ,ax2) , (ax3,ax4)) = plt.subplots( nrows = 2  ,
                                               ncols = 2 ,
                                             figsize=(10,5))
#plotting Individually 
ax1.plot( x , x /2);
ax2.scatter(np.random.random(10) , np.random.random(10));
ax3.bar(list(fruits.keys()) , list(fruits.values()));
ax4.hist(np.random.randint(0,10,2));
````

````python
# option 2 
fig , ax = plt.subplots(nrows = 2 ,
                        ncols = 2 ,
                        figsize= (10, 5))
ax[0,0 ].plot( x , x/2 );
ax[0,1].scatter(np.random.random(10) , np.random.random(10));
ax[1,0].bar(list(fruits.keys()) , list(fruits.values()));
ax[1,1].hist(np.random.randint(0,10,2));
````

````python
#plotting using pandas
car_sales = pd.read_csv(r'E:\conda_project\Pandas\car-sales.csv')
car_sales
````

````python
#removing the dollar size
car_sales['Price'] = car_sales['Price'].str.replace('[\$\,\.]' ,'',regex = True)
car_sales
````

````python
 #removing last two zeros
car_sales['Price'] = car_sales['Price'].str[: -2] #to remove it shld be a string
car_sales
````

- 📅 `car_sales['Sale Date'] = pd.date_range(start='23/10/23', periods=len(car_sales))`: Creates a new column 'Sale Date' in the 'car_sales' dataset, filling it with a sequence of dates starting from '23/10/23' and continuing for the same number of rows as in 'car_sales'.

- 🚗💲 `car_sales`: Displays the updated 'car_sales' dataset, now including the 'Sale Date' column, which helps track when the car sales occurred.

````python
car_sales['Sale Date'] = pd.date_range(start = '23/10/23' , periods = len(car_sales))
car_sales
````

Now we have to add all the pice ,so we can use ```cumsum()```,but before that we need to create the object type to integer type 

````python
car_sales['Total Sale'] = car_sales['Price'].astype(int).cumsum()
car_sales
````

Now plotting the total sales

````python
car_sales.plot( x = 'Sale Date' , y = 'Total Sale');
````

````python
car_sales.plot( x = 'Odometer (KM)' , y = 'Total Sale' , kind = 'scatter');
````

````python
car_sales.plot( x = "Make" , y = "Odometer (KM)" , kind = "bar");
````

**pyplot vs matplotlib Object Oriented**
 * When plotting something quickly , okay to use the pyplot method
 * When plotting advanced ,use object oriented method

    

Using object oriented method 

````python
over_50 = heart_disease[heart_disease['age'] > 50 ]
over_50.head()
````

Marking Age vs cholestrol ,above 50 years of age

````python
over_50.plot(kind = "scatter" , 
             x = 'age' ,
             y = 'chol' , 
             c = 'sex',figsize = (10,6));
````

Using Object Oriented method with pyplot

````python
fig  , ax = plt.subplots(figsize= (10 , 6) )
over_50.plot(  kind = 'scatter' , 
                  x = 'age' ,
                  y = 'chol' ,
                  c = 'sex' ,
                 ax = ax);
ax.set_xlim([60,70]); # changes X-Axis from 20 to 70
````

Using Object oriented method purely

````python
fig , ax = plt.subplots(figsize = (10,6))
#plotting data 
scatter = ax.scatter(x = over_50['age'] ,
                     y = over_50['chol'],
                     c = over_50['target'])
'''
ax.set_xlabel('age')
ax.set_ylabel('chol')
ax.set_title('Age vs Cholestrol') 
or 
''' 
ax.set(title = 'Age vs cholestrol',
       xlabel = 'Age' ,
       ylabel = 'Cholestrol');
#adding legend so we know whats happening 
ax.legend(*scatter.legend_elements() , title = 'Target');

#adding a horizontal line
ax.axhline(over_50['chol'].mean() , linestyle = '-.' );
````

Now lets practice the same thing but with subplots

````python
fig , (ax0,ax1) = plt.subplots( nrows= 2,
                               ncols = 1,
                               figsize = (10 ,10))
#adding data to ax0
scatter = ax0.scatter(  x = over_50['age'],
                        y = over_50['chol'],
                        c = over_50['target']);
#customizing subplot 1
ax0.set( xlabel = 'Age',
         ylabel = 'Cholestrol',
        title = 'Age VS Cholestrol');
#adding legend
ax0.legend(*scatter.legend_elements()  , title = 'Target');
  
    
    
    
#NOW subplot 2 
scatter = ax1.scatter( x = over_50['age'] , 
                       y = over_50['thalach'] ,
                       c = over_50['target']);
#customsizing plot 2 
ax1.set(xlabel = 'Age' ,
        ylabel = 'Thalach(Heart realted)' ,
        title = 'Heart rate wrt age');
#Adding a super title 
fig.suptitle('Heart Disease Analysis' , fontsize = 16 , fontweight = 'bold');
````
